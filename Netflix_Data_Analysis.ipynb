```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for 'type'
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=netflix, hue='type', palette=['blue', 'yellow'], legend=False)
plt.title('Count of TV Shows and Movies')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for 'rating' with rotated labels and different colors
plt.figure(figsize=(13, 8))
sns.countplot(x='rating', hue='rating', data=netflix, palette='Set3', legend=False)
plt.xticks(rotation=90, ha='right')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the count plot with hue
sns.countplot(x='rating', hue='type', data=netflix)

# Set title and show plot
plt.title('Relation between Type and Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
labels = ['Movie', 'TV show']
sizes = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = (0, 0.1)

# Plotting the pie chart
plt.figure(figsize=(9, 9))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')

# Title and legend
plt.title('Distribution of Content Types on Netflix', fontsize=18)
plt.legend(loc='best', fontsize=12)

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the pie chart
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(10, 8))

# Title
plt.title('Distribution of Ratings on Netflix', fontsize=18)

# Display the plot
plt.show()
```
